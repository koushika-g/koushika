```python
import selenium
from selenium import webdriver 
from selenium.webdriver.support.ui import Select
import time
import pandas as pd
import os
from selenium.webdriver.common.by import By
```

```python
driverpath=r"C:\Users\Merit\Downloads\chromedriver_win32 (1)\chromedriver.exe"
driver = webdriver.Chrome(executable_path=driverpath)
driver.maximize_window()
driver.get("https://www.cliffordchance.com/people_and_places.html?_charset_=UTF-8&fname=&lname=&tags=all&tags=all&office=%2Fcontent%2Fcliffordchance%2Fpeople_and_places%2Fpeople%2Foffices%2Fluxembourg&x=64&y=19#partners")
driver.implicitly_wait(10)
time.sleep(5)
driver.find_element(By.XPATH, '//a[text()="OK"]').click()


# 1st category

name=[]
loc=[]
phn=[]
pa=[]
sec=[]




#     print(len(m))
time.sleep(2)

for nav in range(1,5):
    navbar=driver.find_element_by_xpath("//ul[contains(@class,'tabbed tabRegReg greytabs')]//li[{}]".format(nav)).click()
    category=[]
    cat=driver.find_elements_by_xpath("//a[@class='icon_profile']")
    for m in cat:
        links=m.get_attribute("href")
        category.append(links)
    for s in category:
        driver.get(s)
        a=driver.find_elements_by_xpath("//h2[contains(@class,'utils_black')]")
        for i in a:
            name.append(i.text)
            print(name)

        b=driver.find_elements_by_xpath("//h2[contains(text(),'Contact details')]//following::a[1]")
        for j in b:
            loc.append(j.text)
            print(loc)

        c=driver.find_elements_by_xpath("//h2[contains(text(),'Contact details')]//following::li[2]")
        for k in c:
            phn.append(k.text)
            print(phn)

        d=driver.find_elements_by_xpath("//h2[contains(text(),'Contact details')]//following::li[4]")
        for l in d:
            pa.append(l.text)
            print(pa)

        e=driver.find_elements_by_xpath("//span[contains(text(),'Sector')]//following::a[1]")
        for m in e:
            sec.append(m.text)
            print(sec)

        driver.back()
        time.sleep(2)
```

```python
df=pd.DataFrame(list(zip(name,loc,phn,pa,sec)),columns=['name','location','phone number','practice area','sector'])
df
```
